In the following lab we cover some of the classification algorithms previously seen, and evaluate their performance over different dataets. We will understand why different models manage over some datasets while fail over others.

To that end, we will compare logistic regression, decision trees, $k$-NN, SVM and Kernel (Polynomial- and Gaussian) SVM. In brief:
- The logistic regression model aims to model the probability of a sample being classified as $1$ by: $\mathbb{P}\left(y=1\right) = \sigma\left(w^\top x\right)$ where $\sigma$ denotes the sigmoidal function.
- The decision tree model partitions the domain space into a dis-joint union of blocks. Then, prediction is obtained by the block's majoriy vote: $y = \sum y_i \unicode{x1D7D9}\left[x_i\in B\left(x\right)\right]$, where $B\left(x\right)$ denotes the block in which the new sample $x$ belongs to.
- The $k$-NN classifier, a model free classifier, predicts label by the "neighborhood" of a given sample: $\underset{y\in \{0,1\}}{argmax} \sum_{i=1}^k \unicode{x1D7D9}\left[y_{\pi_{i}} = y\right] $. 
- The SVM classifier, learns a separating hyperplane, whereby predictions are made according to the samples location in relation to the hyperplane: $y = sign \left(w^T \cdot \phi\left(x\right)\right) $, where $\phi$ is some transformation of the data. In the case of applying a kernel transformation we would apply
    - A linear kernel: $ K\left(x_i, x_j \right) = x_i^T \cdot x_j $
    - A polynomial kernel: $ K\left(x_i, x_j \right) = \left(1 + \langle x_i, x_j\rangle \right)^k $
    - A Gaussian kernel: $ K\left(x_i, x_j \right) = exp\left(-\frac{1}{2\sigma^2} \left|\left| x_i - x_j \right|\right|^2 \right) $

```python
models = [
    LogisticRegression(),
    DecisionTreeClassifier(max_depth=5), 
    KNeighborsClassifier(n_neighbors=5),
    SVC(kernel='linear', probability=True),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()    
]
```

```python
def plot_all_classifiers(Zs):
    fig = make_subplots(rows=2, cols=3, subplot_titles=("Logistic regression","Desicion Tree", "KNN", "SVM", "LDA", "QDA"))
    
    for i, z in enumerate(Zs):
        fig.add_trace(go.Contour(x=xrange, y=yrange, z=z, colorscale='RdBu', reversescale=True, opacity=0.8), row=(i//3) + 1, col=(i%3)+1)
    fig.update_layout(height=600, width=800, title_text="Decisions bounderies")
    fig.show()

def plot_roc_curves(models, X, y):
    fig = make_subplots(rows=2, cols=3, subplot_titles=("Logistic regression","Desicion Tree", "KNN", "SVM", "LDA", "QDA"))
    
    for i, model in enumerate(models):
        fpr, tpr, th = metrics.roc_curve(y, model.predict_proba(X)[:, 1])
        fig.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines'), row=(i//3) + 1, col=(i%3)+1)
    fig.update_layout(height=600, width=800, title_text="ROC curves", showlegend=False)
    fig.show()
        
        
        
        
```

```python
X = np.random.uniform(low=-5, high=5, size=(1200, 2))

grid_step = 0.2

# Z is the probability grid to create the desicion bounderies
x_min, x_max = np.min(X[:, 0]) - grid_step, np.max(X[:, 0]) + grid_step
y_min, y_max = np.min(X[:, 1]) - grid_step, np.max(X[:, 1]) + grid_step

xrange = np.arange(x_min, x_max, grid_step)
yrange = np.arange(y_min, y_max, grid_step)
xx, yy = np.meshgrid(xrange, yrange)
```

# Generating Data

## Linear data

```python
def even_classes(X, y):
    counts = [np.sum(y == i) for i in [0, 1]]
    min_counts = np.min(counts)
    
    final_X = []
    final_y = []
    for c in [0, 1]:
        final_X.append(X[y == c][:min_counts])
        final_y.append(y[y == c][:min_counts])
    
    final_X = np.array(final_X) 
    final_y = np.array(final_y)

    final_X = final_X.reshape((final_X.shape[0]*final_X.shape[1], final_X.shape[2])) 
    final_y = final_y.reshape((final_y.shape[0]*final_y.shape[1]))

    random_perm = np.random.permutation(final_X.shape[0])
    final_X, final_y = final_X[random_perm], final_y[random_perm]
    return final_X, final_y



def true_linear_labels(X):
    labels = []
    for i in range(X.shape[0]):
        labels.append(1 if (X[i][0]) < (X[i][1]) else 0)
    return np.array(labels)
        
        
y_linear = true_linear_labels(X)
X, y_linear = even_classes(X, y_linear)

go.Figure([go.Scatter(x = X[:, 0], y=X[:, 1], mode = 'markers',
        marker = dict(color=y_linear, colorscale='Portland'))]).show()
```

```python
Zs = []

for model in models:
    model.fit(X, y_linear)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    Zs.append(Z)

plot_all_classifiers(Zs)
plot_roc_curves(models, X, y_linear)
```

## Circular data

```python
#Oracle is the real data classifier - it has the true rule of the classification
def true_circular_labels(X):
    labels = []
    for i in range(X.shape[0]):
        labels.append(1 if (X[i][0])**2  + (X[i][1])**2 <= 3**2 else 0)
    return np.array(labels)
        
        
y_circular = true_circular_labels(X)
X, y_circular = even_classes(X, y_circular)


go.Figure([go.Scatter(x = X[:, 0], y=X[:, 1], mode = 'markers',
        marker = dict(color=y_circular, colorscale='Portland'))]).show()
```

```python
Zs = []

for model in models:
    model.fit(X, y_circular)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    Zs.append(Z)

plot_all_classifiers(Zs)
plot_roc_curves(models, X, y_circular)
```

```python
def true_tree_labels(X):
    labels = []
    for i in range(X.shape[0]):
        labels.append(1 if ((-5 <= X[i][0] <= 0 and -4 <= X[i][1] <= 0) 
                            or (1 <= X[i][0] <= 5 and -1 <= X[i][1] <= 4)) else 0)
    return np.array(labels)

y_tree = true_tree_labels(X)
X, y_tree = even_classes(X, y_tree)


go.Figure([go.Scatter(x = X[:, 0], y=X[:, 1], mode = 'markers',
        marker = dict(color=y_tree, colorscale='Portland'))]).show()
```

```python
Zs = []

for model in models:
    model.fit(X, y_tree)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    Zs.append(Z)

plot_all_classifiers(Zs)
plot_roc_curves(models, X, y_tree)
```

## Gaussian data

```python
def sample_from_gaussians(means, sigmas):
    points, labels = [], []
    for i, (mean, sigma) in enumerate(zip(means, sigmas)):
        x = np.random.multivariate_normal(mean, sigma, size=200)
        points.append(x)
        labels.append(np.zeros((200,)) + i)
    
    X = np.vstack(points)
    y = np.hstack(labels)
    return X, y
    
mean_1 = [-2, -2]
mean_2 = [2, 1]

sigma_1 = [[1, 0], [0, 1]]
sigma_2 = [[1.4, -0.9], [-0.9, 1 ]]
X, y_gaussians = sample_from_gaussians([mean_1, mean_2], [sigma_1, sigma_2])

go.Figure([go.Scatter(x = X[:, 0], y=X[:, 1], mode = 'markers',
                      marker = dict(color=y_gaussians, colorscale='Portland'))]).show()
```

```python
Zs = []

for model in models:
    model.fit(X, y_gaussians)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    Zs.append(Z)

plot_all_classifiers(Zs)
plot_roc_curves(models, X, y_gaussians)
```

Notice:
    
1. Logistic regression and SVM create linear bounderies.

2. Desicion trees make parallel to the axes bouneries.

3. Note how the SVM bouneries are able to reach more complex bounderies as the kernel is more complex 
But, the SVM is basically a linear classifier. That is an importent concept - using mappings and different representations of the data to is also a tool to make the hypothesis class richer.
